```python
sitemap_filename='www.crowdrise.com/sitemap/xml/campaigns/1'
with open(sitemap_filename, 'rb') as f:
    root = xml.etree.ElementTree.fromstring(f.read())

root_tag = root.tag
namespace = root_tag[root_tag.index('{'):root_tag.index('}')+1]

url_data = []
for url in root.findall('.//{}url'.format(namespace)):
    url_data.append(
        dict(
            loc=url.find('./{}loc'.format(namespace)).text,
            lastmod=url.find('./{}lastmod'.format(namespace)).text,
            changefreq=url.find('./{}changefreq'.format(namespace)).text,
            priority=url.find('./{}priority'.format(namespace)).text,
        )
    )
url_data
```
